```python
# interpreter.module_args

'''

Encoding Text Data

'''

class add_datainfo(nlpi):
    
    def __init__(self):
        self.name = 'nlp_encoder'

        # # read config data
        # with open('src/mllibs/corpus/mencoder.json', 'r') as f:
        #     self.json_data = json.load(f)
        #     self.nlp_config = parse_json(self.json_data)

        self.select = None
        self.data = None
        self.args = None

    # verbose output
        
    @staticmethod
    def verbose_set(verbose):
        print(f'set {verbose}')
          
    # set function parameter (related to activation)
            
    @staticmethod
    def sfp(args,preset,key:str):
        
        if(args[key] is not None):
            return eval(args[key])
        else:
            return preset[key] 
        
    # set general parameter
        
    @staticmethod
    def sgp(args,key:str):
        if(args[key] is not None):
            return eval(args[key])
        else:
            return None
           
    # make selection  

    def sel(self,args:dict):
                  
        select = args['pred_task']
        self.data_name = args['data_name']  # name of the data
        self.select = args['pred_task']     # task prediction
        self.data = args['data']
        self.args = args    
        args.name = args['features']
        
        ''' select appropriate predicted method '''
        
        if(self.select == 'add_ac'):
            self.add_ac(self.data,self.args)

    # Add active column
    
    def add_ac(self,name:str):
           
        # if just data is specified
        print(nlpi.data[name]['ac'])
        # nlpi.memory_output.append()

# add_datainfo().add_ac('housing')

lst_join = 'CRIM ZN INDUS CHAS NOX RM AGE MEDV'
lst = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','MEDV']
nlpi.data['housing']['ac']['newest'] = lst
nlpi.data['housing']

# temp = "use active columns columns1"
```
